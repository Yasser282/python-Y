```python
import pandas as pd
import numpy as np

# Create a sample DataFrame for lab values
data = {
    'Date': pd.date_range(start='2023-01-01', periods=20),
    'Test': ['Test1', 'Test2', 'Test1', 'Test3', 'Test2', 'Test3', 'Test1', 'Test2', 'Test3', 'Test1', 'Test2', 'Test3', 'Test1', 'Test2', 'Test3', 'Test1', 'Test2', 'Test3', 'Test1', 'Test2'],
    'Value': np.random.randint(50, 200, size=20),
    'Patient_ID': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10', 'P11', 'P12', 'P13', 'P14', 'P15', 'P16', 'P17', 'P18', 'P19', 'P20']
}

lab_df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(lab_df)

# Group by 'Test' and 'Patient_ID', calculate the total value and average value for each group
grouped_data = lab_df.groupby(['Test', 'Patient_ID']).agg({'Value': ['sum', 'mean']}).reset_index()

# Display the grouped data
print("\nGrouped Data by Test and Patient_ID (sum and mean of Value):")
print(grouped_data)

# Filter rows where the total value is greater than 500
filtered_data = grouped_data[grouped_data[('Value', 'sum')] > 500]

# Display the filtered data
print("\nFiltered Data (Total Value > 500):")
print(filtered_data)

# Create a pivot table to summarize data by 'Test' and 'Patient_ID', showing total value and average value
pivot_table = pd.pivot_table(lab_df, values='Value', index=['Test'], columns=['Patient_ID'], aggfunc={'Value': ['sum', 'mean']}, fill_value=0)

# Display the pivot table
print("\nPivot Table:")
print(pivot_table)
```

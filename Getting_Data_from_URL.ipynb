```python
url = 'https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/Week2/diabetes_unclean.html'

import pandas as pd
import numpy as np


df = pd.read_csv(url)
```

```python
 df.info()
# df.head()
df.describe()
# df.isnull().sum()
```

```python
import numpy as np
from scipy import stats

# Calculate the z-score for each student's height
z = np.abs(stats.zscore(df['HDL']))

# Identify outliers as students with a z-score greater than 3
threshold = 3
outliers = df[z > threshold]

# Print the outliers
print(outliers)
print(z)
```

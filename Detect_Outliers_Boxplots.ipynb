```python
#import the needed packages
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#read the data
df = pd.read_csv('https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/Week2/diabetes_unclean.csv')
# Select only numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Set up the matplotlib figure with subplots
fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(10, 80))

# Create individual boxplots for each numeric column using seaborn
for i, column in enumerate(numeric_columns):
    sns.boxplot(y=df[column], ax=axes[i])
    axes[i].set_title(column)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
